#### Import Section

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

### Figuras y paneles

```python
fig, ax = plt.subplots()
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

ax.plot(x, y)
```

```python
fig, pan_1 = plt.subplots(1, 1)

pan_2 = pan_1.twinx()


x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)

pan_1.plot(x1, y1, 'y')
pan_1.set_xlabel('tiempo')

x2 = np.linspace(0, 4 * np.pi, 100)
y2 = np.cos(x2)

pan_2.plot(x2, y2, 'r')
#pan_2.set_xlabel('tiempo')

plt.savefig('sin_cos_2.png')
```
