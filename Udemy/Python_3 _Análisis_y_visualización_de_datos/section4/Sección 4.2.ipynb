```python
import numpy as np
import pandas as pd
import os
```

```python
CWD = os.getcwd()
BASE_DIR = os.path.dirname(CWD)
RATINGS_PATH = os.path.join(BASE_DIR + '/datasets/ratings.csv'.replace('/', '\\'))
MOVIES_PATH = os.path.join(BASE_DIR + '/datasets/peliculas.csv'.replace('/', '\\'))
USERS_PATH = os.path.join(BASE_DIR + '/datasets/usuarios.csv'.replace('/', '\\'))
```

```python
ratings = pd.read_csv(f'{RATINGS_PATH}')
peliculas = pd.read_csv(f'{MOVIES_PATH}')
usuarios = pd.read_csv(f'{USERS_PATH}')
```

```python
ratings.head()
```

```python
peliculas.head()
```

```python
usuarios.head()
```

```python
peli_ratings = pd.merge(peliculas, ratings)
peli_ratings.head(2)
```

```python
clasif = pd.merge(peli_ratings, usuarios)
clasif.head()
```

```python
mas_ratings = clasif.groupby(by = 'titulo').size().sort_values(ascending=False)[:25]
mas_ratings
```

```python
peli_stats = clasif.groupby('titulo').agg({'rating':[np.size, np.mean]})
peli_stats.sort_values([('rating', 'size')], ascending = False)
```

```python
rate = 100
PelisOverRate= peli_stats[peli_stats[('rating', 'size')] > rate].sort_values([('rating', 'mean')], ascending = False)
print(PelisOverRate.head(15))
print(f'El número de peliculas por encima de {rate} ratings es {PelisOverRate.shape}')
```
